```python
import pygame
from collections import deque
from time import time
class button():
    def __init__(self,x,y,image,scale):
        width=image.get_width()
        height=image.get_height()
        self.image=pygame.transform.scale(image,(int(width*scale),int(height*scale)))
        self.rect=self.image.get_rect()
        self.rect.topleft=(x,y)
        self.clicked=False
    def draw(self,screen):
        action=False
        #get mouse position
        pos=pygame.mouse.get_pos()
        #check mouseover and clicked conditions
        if self.rect.collidepoint(pos):
            if pygame.mouse.get_pressed()[0]==1 and self.clicked==False:
                self.clicked=True
                action=True
            
            if pygame.mouse.get_pressed()[0]==0:
                self.clicked=False
        
        #draw button on screen
        screen.blit(self.image,(self.rect.x,self.rect.y))
        return action
    
def draw_text(text,font,text_col,x,y):
    img=font.render(text,True,text_col)
    screen.blit(img,(x,y))
        
    

pygame.init()

screen=pygame.display.set_mode((800,600))
pygame.display.set_caption('TOWER OF HANOI')

#game variables
moving1=False
moving2=False
moving3=False
moved=False
stack=deque()
stack1=deque()
stack2=deque()
stack3=deque()
stack_col=deque()
clicked=False
h1=0
h2=0
h3=0
move=0
menu_state='main'
start_time=0
end_time=0
game_paused=False

#loading the image
start_img=pygame.image.load('button_start.png').convert_alpha()
quit_img=pygame.image.load('button_quit.png').convert_alpha()
easy_img=pygame.image.load('easy.png').convert_alpha()
medium_img=pygame.image.load('medium.png').convert_alpha()
hard_img=pygame.image.load('hard.png').convert_alpha()
back_img=pygame.image.load('button_back.png').convert_alpha()
resume_img=pygame.image.load('button_resume.png').convert_alpha()
win_img=pygame.image.load('win.png').convert_alpha()
width=win_img.get_width()
height=win_img.get_height()
win_img=pygame.transform.scale(win_img,(int(width*0.75),int(height*1)))

#create button instances
start_button=button(325,75,start_img,0.3)
quit_button=button(340,325,quit_img,1)
easy_button=button(325,75,easy_img,0.05)
medium_button=button(290,200,medium_img,0.17)
hard_button=button(325,360,hard_img,0.05)
back_button=button(600,5,back_img,0.9)
resume_button=button(320,75,resume_img,1)
win_rect=win_img.get_rect()


#bead height
bh=20

#coordinates of beads
rectbead1=pygame.Rect(130,455,150,bh)
rectbead2=pygame.Rect(140,435,130,bh)
rectbead3=pygame.Rect(150,415,110,bh)
rectbead4=pygame.Rect(160,395,90,bh)
rectbead5=pygame.Rect(170,375,70,bh)


stack.append(rectbead1)
stack.append(rectbead2)
stack.append(rectbead3)
stack.append(rectbead4)
stack.append(rectbead5)

stack_col.append((205, 133, 63, 255))
stack_col.append((205, 105, 201, 255))
stack_col.append((119, 136, 153, 255))
stack_col.append((218, 165, 32, 255))
stack_col.append((180, 238, 180, 255))

#rect of containers
rect_cont1=pygame.Rect(115,330,175,150)
rect_cont2=pygame.Rect(330,330,175,150)
rect_cont3=pygame.Rect(545,330,175,150)

rect_cont=[rect_cont1,rect_cont2,rect_cont3]

#rect of holders
rect_hold1=pygame.Rect(200,330,10,150)
rect_hold2=pygame.Rect(415,330,10,150)
rect_hold3=pygame.Rect(630,330,10,150)

rect_hold=[rect_hold1,rect_hold2,rect_hold3]


run=True
while run:
    screen.fill((100,100,100))
    for event in pygame.event.get():
        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_ESCAPE:
                game_paused=True
                menu_state='paused'
        if event.type==pygame.QUIT:
            run=False
        
    
    #creaating menus
        
    if menu_state=='main':
        rectbead5=pygame.Rect(170,375,70,bh)
        if start_button.draw(screen):
            menu_state='select_level'
        elif quit_button.draw(screen):
            run=False
            
    elif menu_state=='select_level':

        screen.fill((200,0,0))
        tim=0
        if easy_button.draw(screen):
            menu_state='in_game'
            n=3
            h1=n*20
            stack1.append(rectbead1)
            stack1.append(rectbead2)
            stack1.append(rectbead3)
        if medium_button.draw(screen):
            menu_state='in_game'
            n=4
            h1=n*20
            stack1.append(rectbead1)
            stack1.append(rectbead2)
            stack1.append(rectbead3)
            stack1.append(rectbead4)
        if hard_button.draw(screen):
            menu_state='in_game'
            n=5
            h1=n*20
            stack1.append(rectbead1)
            stack1.append(rectbead2)
            stack1.append(rectbead3)
            stack1.append(rectbead4)
            stack1.append(rectbead5)
        if back_button.draw(screen):
            menu_state='main'
        
    
            
    #pause menu
    elif menu_state=='paused':
        if resume_button.draw(screen):
            menu_state='in_game'
            game_paused=False
        if back_button.draw(screen):
            menu_state='main'
        if quit_button.draw(screen):
            run=False
            
          
    elif menu_state=='in_game':
        screen.fill((0,0,0))
        draw_text('moves : '+str(move),font,(255,255,255),0,10)
        if (tim==0):
            start_time=time()
            tim+=1
        star_time=time()
        draw_text('time : '+str(round(star_time-start_time,5)),font,(255,255,255),400,10)


        #drawing the table
        pygame.draw.rect(screen,(100,40,0),(0,450,800,150))


        #drawing the holders
        pygame.draw.rect(screen,(150,75,0),(115,475,175,15),border_radius=10)
        pygame.draw.rect(screen,(150,75,0),(330,475,175,15),border_radius=10)
        pygame.draw.rect(screen,(150,75,0),(545,475,175,15),border_radius=10)
        pygame.draw.rect(screen,(150,75,0),(200,330,10,150),border_radius=10)
        pygame.draw.rect(screen,(150,75,0),(415,330,10,150),border_radius=10)
        pygame.draw.rect(screen,(150,75,0),(630,330,10,150),border_radius=10)
        
        for i in range(0,n):
            pygame.draw.rect(screen,stack_col[i],stack[n-i-1],border_radius=10)
        


        pos=pygame.mouse.get_pos()
        
        if (len(stack1)==0 and len(stack2)==0):
            end_time=time()
            times=round(end_time-start_time,5)
            menu_state='game_won'
            
    
        #shall not move the bead when clicked to move
        if moved==True:
            if pygame.mouse.get_pressed()[0]==1 and clicked==False:
                clicked=True
                continue
            if pygame.mouse.get_pressed()[0]==0:
                moved=False
                clicked=False
        else:
            # selecting bead from 1st container
            if rect_cont[0].collidepoint(pos):
                if (len(stack1)!=0):
                    if(moving2==False and moving3==False and moved==False):
                        if pygame.mouse.get_pressed()[0]==1 and clicked==False:
                            clicked=True
                            moving1=True
                        if pygame.mouse.get_pressed()[0]==0:
                            clicked=False
            #selecting bead from 2nd container        
            if rect_cont[1].collidepoint(pos):
                if (len(stack2)!=0):
                    if(moving1==False and moving3==False and moved==False):
                        if pygame.mouse.get_pressed()[0]==1 and clicked==False:
                            clicked=True
                            moving2=True
                        if pygame.mouse.get_pressed()[0]==0:
                            clicked=False

            #selecting bead from 3rd container        
            if rect_cont[2].collidepoint(pos):
                if (len(stack3)!=0):
                    if(moving1==False and moving2==False and moved==False):
                        if pygame.mouse.get_pressed()[0]==1 and clicked==False:
                            clicked=True
                            moving3=True
                        if pygame.mouse.get_pressed()[0]==0:
                            clicked=False


        #moving the beads from 1st container     
        if moving1:
            if(len(stack1)!=0):
                keys=pygame.key.get_pressed()
                stack1[len(stack1)-1].center=pos
                #moving the beads from 1st to 2nd container
                if rect_hold[1].colliderect(stack1[len(stack1)-1]):
                    if keys[pygame.K_LSHIFT]:
                            if (len(stack2)==0):
                                    stack1[len(stack1)-1].center=(420,465-h2)
                                    stack2.append(stack1[len(stack1)-1])
                                    h2+=20
                                    h1-=20
                                    move+=1
                                    stack1.pop()
                                    moving1=False
                                    moved=True
                            else:
                                    if(stack1[len(stack1)-1].width>stack2[len(stack2)-1].width ):
                                        continue
                                    else:
                                        stack1[len(stack1)-1].center=(420,465-h2)
                                        stack2.append(stack1[len(stack1)-1])
                                        h2+=20
                                        h1-=20
                                        move+=1
                                        stack1.pop()
                                        moving1=False
                                        moved=True

                #moving the bead from 1st to 3rd container
                if(len(stack1)!=0):
                    if rect_hold[2].colliderect(stack1[len(stack1)-1]):
                        if keys[pygame.K_LSHIFT]:
                            if (len(stack3)==0):
                                    stack1[len(stack1)-1].center=(635,465-h3)
                                    stack3.append(stack1[-1])
                                    h3+=20
                                    h1-=20
                                    move+=1
                                    stack1.pop()
                                    moving1=False
                                    moved=True
                            else:
                                    if(stack1[len(stack1)-1].width>stack3[len(stack3)-1].width ):
                                        continue
                                    else:
                                        stack1[len(stack1)-1].center=(635,465-h3)
                                        stack3.append(stack1[len(stack1)-1])
                                        h3+=20
                                        h1-=20
                                        move+=1
                                        stack1.pop()
                                        moving1=False
                                        moved=True

                #moving from 1st to 1st container
                if (len(stack1)!=0):
                    if rect_hold[0].colliderect(stack1[len(stack1)-1]):
                        if keys[pygame.K_LSHIFT]:
                            stack1[len(stack1)-1].center=(205,465-(h1-20))
                            moving1=False
                            moved=True
                else:
                    continue



        #moving from 2nd container              
        if moving2:
            if(len(stack2)!=0):
                keys=pygame.key.get_pressed()
                stack2[len(stack2)-1].center=pos
                #moving the beads from 2nd to 1st container
                if rect_hold[0].colliderect(stack2[len(stack2)-1]):
                    if keys[pygame.K_LSHIFT]:
                        if (len(stack1)==0):
                                stack2[len(stack2)-1].center=(205,465-h1)
                                h1+=20
                                h2-=20
                                move+=1
                                stack1.append(stack2[len(stack2)-1])
                                stack2.pop()
                                moving2=False
                                moved=True
                        else:
                                if(stack2[len(stack2)-1].width>stack1[len(stack1)-1].width):
                                    continue

                                else:
                                    stack2[len(stack2)-1].center=(205,465-h1)
                                    h1+=20
                                    h2-=20
                                    move+=1
                                    stack1.append(stack2[len(stack2)-1])
                                    stack2.pop()
                                    moving2=False
                                    moved=True  


                #moving the beads from 2nd to 3rd container
                if(len(stack2)!=0):
                    if rect_hold[2].colliderect(stack2[len(stack2)-1]):
                        if keys[pygame.K_LSHIFT]:
                            if (len(stack3)==0):
                                    stack2[len(stack2)-1].center=(635,465-h3)
                                    h3+=20
                                    h2-=20
                                    move+=1
                                    stack3.append(stack2[len(stack2)-1])
                                    stack2.pop()
                                    moving2=False
                                    moved=True

                            else:
                                    if(stack2[len(stack2)-1].width>stack3[len(stack3)-1].width):
                                        continue

                                    else:
                                        stack2[len(stack2)-1].center=(635,465-h3)
                                        h3+=20
                                        h2-=20
                                        move+=1
                                        stack3.append(stack2[len(stack2)-1])
                                        stack2.pop()
                                        moving2=False
                                        moved=True
                #moving from 2nd to 2nd container
                if (len(stack2)!=0):
                    if rect_hold[1].colliderect(stack2[len(stack2)-1]):
                        if keys[pygame.K_LSHIFT]:
                            stack2[len(stack2)-1].center=(420,465-(h2-20))
                            moving2=False
                            moved=True
                else:
                    continue



        #moving from 3rd container
        if moving3:
            if(len(stack3)!=0):
                keys=pygame.key.get_pressed()
                stack3[len(stack3)-1].center=pos
                #moving the beads from 3rd to 1st container
                if rect_hold[0].colliderect(stack3[len(stack3)-1]):
                    if keys[pygame.K_LSHIFT]:
                        if (len(stack1)==0):
                                stack3[len(stack3)-1].center=(205,465-h1)
                                h1+=20
                                h3-=20
                                move+=1
                                stack1.append(stack3[len(stack3)-1])
                                stack3.pop()
                                moving3=False
                                moved=True


                        else:
                                if(stack3[len(stack3)-1].width>stack1[len(stack1)-1].width):
                                    continue

                                else:
                                    stack3[len(stack3)-1].center=(205,465-h1)
                                    h1+=20
                                    h3-=20
                                    move+=1
                                    stack1.append(stack3[len(stack3)-1])
                                    stack3.pop()
                                    moving3=False
                                    moved=True  

                #moving the beads from 3rd to 2nd container
                if(len(stack3)!=0):
                    if rect_hold[1].colliderect(stack3[len(stack3)-1]):
                        if keys[pygame.K_LSHIFT]:
                            if (len(stack2)==0):
                                    stack3[len(stack3)-1].center=(420,465-h2)
                                    h2+=20
                                    h3-=20
                                    move+=1
                                    stack2.append(stack3[len(stack3)-1])
                                    stack3.pop()
                                    moving3=False
                                    moved=True

                            else:
                                    if(stack3[len(stack3)-1].width>stack2[len(stack2)-1].width):
                                        continue

                                    else:
                                        stack3[len(stack3)-1].center=(420,465-h2)
                                        h2+=20
                                        h3-=20
                                        move+=1
                                        stack2.append(stack3[len(stack3)-1])
                                        stack3.pop()
                                        moving3=False
                                        moved=True
                #moving from 3rd to 3rd container
                if (len(stack3)!=0):
                    if rect_hold[2].colliderect(stack3[len(stack3)-1]):
                        if keys[pygame.K_LSHIFT]:
                            stack3[len(stack3)-1].center=(635,465-(h3-20))
                            moving3=False
                            moved=True
                else:
                    continue
                    

                    
                    
                    
    #define fonts
    font=pygame.font.SysFont('arialblack',40)
    
    #game won
    if menu_state=='game_won':
        h3=0
        stack.clear()
        stack3.clear()
        if (len(stack)==0):
            rectbead1=pygame.Rect(130,455,150,bh)
            rectbead2=pygame.Rect(140,435,130,bh)
            rectbead3=pygame.Rect(150,415,110,bh)
            rectbead4=pygame.Rect(160,395,90,bh)
            rectbead5=pygame.Rect(170,375,70,bh)
        screen.fill((0,0,0))
        screen.blit(win_img,win_rect)
        draw_text('moves : '+str(move),font,(255,255,255),0,400)
        draw_text('time : '+str(times),font,(255,255,255),300,400)
        draw_text('you won. ',font,(255,255,255),0,450)
        draw_text('press n to quit or y to play again',font,(255,255,255),0,500)
        keys=pygame.key.get_pressed()

        if keys[pygame.K_n]:
            #exit the loops
            run=False
        if keys[pygame.K_y]:
            #go to select level menu
            menu_state='select_level'
            tim=0
            move=0
            stack.append(rectbead1)
            stack.append(rectbead2)
            stack.append(rectbead3)
            stack.append(rectbead4)
            stack.append(rectbead5)



                    
        
            
            
    pygame.display.update()
    
pygame.quit()
```
